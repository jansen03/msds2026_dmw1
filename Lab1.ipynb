## Phase 1: Data Cleaning, Validation, and Benchmark Establishment

### 1. Data Ingestion and Cleaning

```python
conn = sqlite3.connect('anime.db')
cursor = conn.cursor()
command = """SELECT * FROM anime_table
where genre is not '' AND genre IS NOT NULL AND
type IS NOT '' AND type IS NOT NULL AND
episodes IS NOT 'Unknown' AND episodes IS NOT NULL AND
rating IS NOT NULL

"""
anime_clean_df = pd.read_sql(command, conn)
anime_clean_df
```

```python
cursor = conn.cursor()
command = """SELECT * FROM anime_genre
where genre is not '' AND genre IS NOT NULL AND
type IS NOT '' AND type IS NOT NULL AND
episodes IS NOT 'Unknown' AND episodes IS NOT NULL AND
rating IS NOT NULL

"""
anime_genre_clean_df = pd.read_sql(command, conn)
anime_genre_clean_df
```

```python
command = """SELECT * FROM rating_table
WHERE rating IS NOT '-1'

"""
rating_clean_df = pd.read_sql(command, conn)
rating_clean_df
```

### 2. Rating System Validation:

#### a. Standard Deviation for Each Anime

```python
consensus_metrics = rating_clean_df.groupby('anime_id')['rating'].agg(
    std_dev='std',
    num_ratings='size'
).reset_index()
consensus_metrics
```

```python
final_consensus_df = anime_clean_df.merge(consensus_metrics, on='anime_id', how='inner')
final_consensus_df
```

#### b. User Engagement (Rating Completion Rate)

```python
total_entries = rating_df.shape[0]

completed_ratings = rating_df[rating_df['rating'] != -1].shape[0]

rating_completion_rate = completed_ratings / total_entries

print(f"Total 'Watched' Entries: {total_entries:,}")
print(f"Entries with Numerical Scores: {completed_ratings:,}")
print(f"Rating Completion Rate: {rating_completion_rate:.2%}")
```

# Performance Benchmark
